### Pivot Tables

```python
import pandas as pd
import numpy as n
```

```python
df = pd.read_csv(r"C:\Users\Admin\Documents\Course Files\QADHPYTHON\data\loan_data.csv")
```

```python
df
```

```python
#how the average income changed with default or not
```

```python
df[['Income', 'Default', 'Term']].groupby(['Default', 'Term']).mean().round()
```

```python
df.pivot_table(values = ['Income', 'Balance'], index = ['Default', df['Balance']>10000], columns = ['Term']).round()
```

```python
df.pivot_table(values = ['Income', 'Balance'], index = ['Default', df['Balance'].map(lambda bal: 'Low' if bal<1000 else 'High')], columns = ['Term']).round()
```
